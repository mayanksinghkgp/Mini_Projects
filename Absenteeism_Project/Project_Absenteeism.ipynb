```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('Absenteeism_data.csv')
```

```python
df = raw_data.copy()
```

```python
df.head()
```

```python
df = df.drop('ID', axis = 1)
```

```python
def ROA(x):
    if x == 0:
        return 0
    elif x<=14:
        return 1
    elif x>14 and x<=17:
        return 2
    elif x>17 and x<=21:
        return 3
    elif x>=22:
        return 4
```

```python
dfr = df['Reason for Absence'].map(ROA)
dfr
```

```python
dfr = pd.get_dummies(dfr, drop_first = True)
```

```python
dfr
```

```python
df = pd.concat([df,dfr], axis = 1)
df = df.drop('Reason for Absence', axis = 1)
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
```

```python
df.head()
```

```python
column_names_reoder = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reoder]
```

```python
df.head()
```
